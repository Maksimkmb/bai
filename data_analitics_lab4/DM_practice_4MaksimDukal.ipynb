#### **Система Лоренца:**

Система Лоренца описується трьома диференціальними рівняннями:

$$
\begin{cases}
\frac{dx}{dt} = \sigma (y - x) \\
\frac{dy}{dt} = x (\rho - z) - y \\
\frac{dz}{dt} = xy - \beta z
\end{cases}
$$

де:
- $( \sigma )$ — коефіцієнт Прандтля (зазвичай $ \sigma = 10 $),
- $( \rho )$ — числовий параметр для турбулентності (зазвичай $ \rho = 28 $),
- $( \beta )$ — коефіцієнт відношення висоти до довжини (зазвичай $ \beta = 8/3 $).

#### **Об'єкти Нарендра-Пархтізаратхі:**

Об'єкти Нарендра-Пархтізаратхі описуються наступними рівняннями:

$$
\begin{cases}
x_1(t+1) = \alpha_1 x_1(t) + \beta_1 x_2(t) + \gamma_1 x_3(t) + \delta_1 \\
x_2(t+1) = \alpha_2 x_1(t) + \beta_2 x_2(t) + \gamma_2 x_3(t) + \delta_2 \\
x_3(t+1) = \alpha_3 x_1(t) + \beta_3 x_2(t) + \gamma_3 x_3(t) + \delta_3
\end{cases}
$$

де $(\alpha_i)$, $(\beta_i)$, $(\gamma_i)$ і $(\delta_i)$ — коефіцієнти для кожного з рівнянь.

### **Нелінійних рівнянь Нарендра-Партхізаратхі:**

\dot{x}_1 = -a_1 x_1 + b_1 x_2^2

$$
\dot{x}_2 = -a_2 x_2 + b_2 \sin(x_1)

\dot{x}_1 = a_1 \sin(x_2) - b_1 x_1^2

$$
\dot{x}_2 = a_2 x_1 x_2 - b_2 \cos(x_1)

$$
\dot{x}_2 = a_2 \sin(x_1) - b_2 x_2

6. $$

$$
\dot{x}_2 = a_2 x_1 - b_2 \sin(x_2)

\dot{x}_1 = a_1 \sin(x_2) - b_1 x_1^2

### Параметри:
- $( a_1, a_2 )$, $( b_1, b_2 )$ — це коефіцієнти, які можуть бути варійовані для дослідження динаміки системи та впливу на прогнозування.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import Ridge
from scipy.integrate import odeint

```

```python
sigma = 10
rho = 28
beta = 8/3

def lorenz_system(x, y, z, sigma, rho, beta):
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return dx, dy, dz

x = 1.0
y = 1.0
z = 1.0

dt = 0.01
num_steps = 10000

xs = np.empty(num_steps)
ys = np.empty(num_steps)
zs = np.empty(num_steps)

xs[0], ys[0], zs[0] = x, y, z

for i in range(1, num_steps):
    dx, dy, dz = lorenz_system(x, y, z, sigma, rho, beta)
    x += dx * dt
    y += dy * dt
    z += dz * dt
    xs[i], ys[i], zs[i] = x, y, z

df = pd.DataFrame({'x': xs, 'y': ys, 'z': zs})

df.to_csv('lorenz_system_4lab.csv', index=False)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c=range(num_steps), cmap='viridis', s=1)
ax.set_title("lorenz")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
plt.show()

```

```python
alpha_1 = 1.0
alpha_2 = 1.5
beta_1 = 0.5
beta_2 = 0.3
gamma_1 = 2.0
gamma_2 = 0.8
delta_1 = 0.2
delta_2 = 0.5


def narendra_parthasarathy_system(x, y, z, w, alpha_1, alpha_2, beta_1, beta_2, gamma_1, gamma_2, delta_1, delta_2):
    dx = alpha_1 * y - beta_1 * x * z
    dy = alpha_2 * x * z - beta_2 * y
    dz = gamma_1 * x - delta_1 * y * z
    dw = gamma_2 * y * z - delta_2 * w
    return dx, dy, dz, dw

x = 1.0
y = 0.5
z = 2.0
w = 0.5

dt = 0.0007
num_steps = 10000

xs = np.empty(num_steps)
ys = np.empty(num_steps)
zs = np.empty(num_steps)
ws = np.empty(num_steps)

xs[0], ys[0], zs[0], ws[0] = x, y, z, w

for i in range(1, num_steps):
    dx, dy, dz, dw = narendra_parthasarathy_system(x, y, z, w, alpha_1, alpha_2, beta_1, beta_2, gamma_1, gamma_2, delta_1, delta_2)
    x += dx * dt
    y += dy * dt
    z += dz * dt
    w += dw * dt
    xs[i], ys[i], zs[i], ws[i] = x, y, z, w

df = pd.DataFrame({'x': xs, 'y': ys, 'z': zs, 'w': ws})

df.to_csv('narendra_parthasarathy_system_4lab.csv', index=False)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs, ys, zs, c=range(num_steps), cmap='plasma', s=1)
ax.set_title("narendra_parthasarathy")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()

```

```python
a1 = 1.0
a2 = 1.0
b1 = 0.5
b2 = 0.5

def system(state, z, a1, a2, b1, b2):
    x, y = state
    dx_dt = a1 * np.sin(y) - b1 * x**2
    dy_dt = a2 * x * y - b2 * np.cos(x)
    return [dx_dt, dy_dt]

initial_state = [1.0, 1.0]

z = np.linspace(0, 10, 1000)

solution = odeint(system, initial_state, z, args=(a1, a2, b1, b2))

x = solution[:, 0]
y = solution[:, 1]

df = pd.DataFrame({
    'z': z,
    'x': x,
    'y': y
})

df.to_csv('custom_dataset_4lab.csv', index=False)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
ax.set_xlabel('X (x1)')
ax.set_ylabel('Y (x2)')
ax.set_zlabel('Z (time)')
plt.show()
```

```python
df1_lorenz_system = pd.read_csv('lorenz_system_4lab.csv')
df2_narendra_parthasarathy = pd.read_csv('narendra_parthasarathy_system_4lab.csv')
df3_custom_date = pd.read_csv('custom_dataset_4lab.csv')


X_lorenz = df1_lorenz_system[['x', 'y', 'z']]
X_narendra = df2_narendra_parthasarathy[['x', 'y', 'z']]
X_custom = df3_custom_date[['x', 'y', 'z']]


X_train_lorenz, X_temp_lorenz = train_test_split(X_lorenz, test_size=0.3, random_state=1)
X_val_lorenz, X_test_lorenz = train_test_split(X_temp_lorenz, test_size=0.5, random_state=1)

X_train_narendra, X_temp_narendra = train_test_split(X_narendra, test_size=0.3, random_state=1)
X_val_narendra, X_test_narendra = train_test_split(X_temp_narendra, test_size=0.5, random_state=1)


X_train_custom, X_temp_custom = train_test_split(X_custom, test_size=0.3, random_state=1)
X_val_custom, X_test_custom = train_test_split(X_temp_custom, test_size=0.5, random_state=1)


scaler = StandardScaler()

X_train_lorenz_scaled = scaler.fit_transform(X_train_lorenz)
X_val_lorenz_scaled = scaler.transform(X_val_lorenz)
X_test_lorenz_scaled = scaler.transform(X_test_lorenz)

X_train_narendra_scaled = scaler.fit_transform(X_train_narendra)
X_val_narendra_scaled = scaler.transform(X_val_narendra)
X_test_narendra_scaled = scaler.transform(X_test_narendra)

X_train_custom_scaled = scaler.fit_transform(X_train_custom)
X_val_custom_scaled = scaler.transform(X_val_custom)
X_test_custom_scaled = scaler.transform(X_test_custom)

```

```python
linear_regression = LinearRegression()
decision_tree = DecisionTreeRegressor(random_state=1)
random_forest = RandomForestRegressor(random_state=1)
svm = SVR()

def train_and_evaluate_model(model, X_train, X_val, y_train, y_val):
    model.fit(X_train, y_train)
    
    y_val_pred = model.predict(X_val)
    mse = mean_squared_error(y_val, y_val_pred)
    
    print(f"Mean Squared Error {model.__class__.__name__}: {mse:.4f}")

y_train_lorenz = X_train_lorenz_scaled[:, 0]
y_val_lorenz = X_val_lorenz_scaled[:, 0]
y_test_lorenz = X_test_lorenz_scaled[:, 0]

y_train_narendra = X_train_narendra_scaled[:, 0]
y_val_narendra = X_val_narendra_scaled[:, 0]
y_test_narendra = X_test_narendra_scaled[:, 0]

y_train_custom = X_train_custom_scaled[:, 0]
y_val_custom = X_val_custom_scaled[:, 0]
y_test_custom = X_test_custom_scaled[:, 0]

print("Lorenz system result:")
for model in [linear_regression, decision_tree, random_forest, svm]:
    train_and_evaluate_model(model, X_train_lorenz_scaled, X_val_lorenz_scaled, y_train_lorenz, y_val_lorenz)

print("\nNarendra-Parthasarathy system result:")
for model in [linear_regression, decision_tree, random_forest, svm]:
    train_and_evaluate_model(model, X_train_narendra_scaled, X_val_narendra_scaled, y_train_narendra, y_val_narendra)

print("\nCustum system result:")
for model in [linear_regression, decision_tree, random_forest, svm]:
    train_and_evaluate_model(model, X_train_custom_scaled, X_val_custom_scaled, y_train_custom, y_val_custom)
```

```python

param_grid_linear = {
    'alpha': [0.1, 1.0, 10.0]  
}

param_grid_tree = {
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

param_grid_forest = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

param_grid_svm = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}

def grid_search_model(model, param_grid, X_train, y_train):
    grid_search = GridSearchCV(model, param_grid, scoring='neg_mean_squared_error', cv=3, n_jobs=-1)
    grid_search.fit(X_train, y_train)
    
    print(f"\nbest param {model.__class__.__name__}: {grid_search.best_params_}")
    return grid_search.best_estimator_

print("Grid Search for Lorenz system:")

ridge = Ridge()
best_ridge_lorenz = grid_search_model(ridge, param_grid_linear, X_train_lorenz_scaled, y_train_lorenz)

best_tree_lorenz = grid_search_model(decision_tree, param_grid_tree, X_train_lorenz_scaled, y_train_lorenz)

best_forest_lorenz = grid_search_model(random_forest, param_grid_forest, X_train_lorenz_scaled, y_train_lorenz)

best_svm_lorenz = grid_search_model(svm, param_grid_svm, X_train_lorenz_scaled, y_train_lorenz)

print("\nGrid Search for Narendra-Parthasarathy system:")

best_ridge_narendra = grid_search_model(ridge, param_grid_linear, X_train_narendra_scaled, y_train_narendra)

best_tree_narendra = grid_search_model(decision_tree, param_grid_tree, X_train_narendra_scaled, y_train_narendra)

best_forest_narendra = grid_search_model(random_forest, param_grid_forest, X_train_narendra_scaled, y_train_narendra)

best_svm_narendra = grid_search_model(svm, param_grid_svm, X_train_narendra_scaled, y_train_narendra)

print("\nGrid Search for custom system:")

best_ridge_custom = grid_search_model(ridge, param_grid_linear, X_train_custom_scaled, y_train_custom)

best_tree_custom = grid_search_model(decision_tree, param_grid_tree, X_train_custom_scaled, y_train_custom)

best_forest_custom = grid_search_model(random_forest, param_grid_forest, X_train_custom_scaled, y_train_custom)

best_svm_custom = grid_search_model(svm, param_grid_svm, X_train_custom_scaled, y_train_custom)
```

```python
def train_and_evaluate_model(model, X_train, X_val, y_train, y_val, X_test, y_test):
    model.fit(X_train, y_train)
    
    y_val_pred = model.predict(X_val)
    val_mse = mean_squared_error(y_val, y_val_pred)
    val_r2 = r2_score(y_val, y_val_pred)
    
    print(f"{model.__class__.__name__} - Validation MSE: {val_mse:.4f}, R²: {val_r2:.4f}")
    
    y_test_pred = model.predict(X_test)
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    
    print(f"{model.__class__.__name__} - Test MSE: {test_mse:.4f}, R²: {test_r2:.4f}")

print("Lorenz system results:")
for model in [linear_regression, decision_tree, random_forest, svm]:
    train_and_evaluate_model(model, X_train_lorenz_scaled, X_val_lorenz_scaled, y_train_lorenz, y_val_lorenz, X_test_lorenz_scaled, y_test_lorenz)

print("\nNarendra-Parthasarathy system results:")
for model in [linear_regression, decision_tree, random_forest, svm]:
    train_and_evaluate_model(model, X_train_narendra_scaled, X_val_narendra_scaled, y_train_narendra, y_val_narendra, X_test_narendra_scaled, y_test_narendra)

print("\nCustom system results:")
for model in [linear_regression, decision_tree, random_forest, svm]:
    train_and_evaluate_model(model, X_train_custom_scaled, X_val_custom_scaled, y_train_custom, y_val_custom, X_test_custom_scaled, y_test_custom)

```

```python
def compare_models(models, X_train, X_val, y_train, y_val, X_test, y_test):
    results = []
    for model in models:
        model.fit(X_train, y_train)
        
        y_val_pred = model.predict(X_val)
        val_mse = mean_squared_error(y_val, y_val_pred)
        val_r2 = r2_score(y_val, y_val_pred)
        
        y_test_pred = model.predict(X_test)
        test_mse = mean_squared_error(y_test, y_test_pred)
        test_r2 = r2_score(y_test, y_test_pred)
        
        results.append({
            'Model': model.__class__.__name__,
            'Validation MSE': val_mse,
            'Validation R²': val_r2,
            'Test MSE': test_mse,
            'Test R²': test_r2
        })
    
    return pd.DataFrame(results)

models = [linear_regression, decision_tree, random_forest, svm]

print("Lorenz system comparison:")
df_lorenz = compare_models(models, X_train_lorenz_scaled, X_val_lorenz_scaled, y_train_lorenz, y_val_lorenz, X_test_lorenz_scaled, y_test_lorenz)
print(df_lorenz)

print("\nNarendra-Parthasarathy system comparison:")
df_narendra = compare_models(models, X_train_narendra_scaled, X_val_narendra_scaled, y_train_narendra, y_val_narendra, X_test_narendra_scaled, y_test_narendra)
print(df_narendra)

print("\nCustom system comparison:")
df_custom = compare_models(models, X_train_custom_scaled, X_val_custom_scaled, y_train_custom, y_val_custom, X_test_custom_scaled, y_test_custom)
print(df_custom)
```

```python
def plot_predictions(models, X_test, y_test, system_name):
    plt.figure(figsize=(20, 8))
    
    for model in models:
        y_pred = model.predict(X_test)
        plt.plot(y_test, label='Actual', color='black', linestyle='dashed', linewidth=2)
        plt.plot(y_pred, label=f'{model.__class__.__name__} Predicted', alpha=0.7)
        
        plt.title(f'{system_name}: Actual vs Predicted for {model.__class__.__name__}')
        plt.xlabel('Sample Index')
        plt.ylabel('Value')
        plt.legend()
        plt.show()

print("Lorenz system predictions:")
plot_predictions(models, X_test_lorenz_scaled, y_test_lorenz, 'Lorenz system')

print("\nNarendra-Parthasarathy system predictions:")
plot_predictions(models, X_test_narendra_scaled, y_test_narendra, 'Narendra-Parthasarathy system')

print("\nCustom system predictions:")
plot_predictions(models, X_test_custom_scaled, y_test_custom, 'Custom system')
```
